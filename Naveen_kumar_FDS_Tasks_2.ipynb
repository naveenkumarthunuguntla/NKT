```python
#Simple Data Analysis
list = [4,5,2,3,9,8,7,1,10,6]

#Calculate the Average
def average(list):
    sum = 0
    for i in list:
        sum += i
        result = sum/len(list)
    return result


#Find the minimum value
def minimum(list):
    min = list[0]
    for i in list:
        if i < min:
            min = i
        else:
            pass
    return min


#Find the maximum value
def maximum(list):
    min = list[0]
    for i in list:
        if i > min:
            min = i
        else:
            pass
    return min
print(average(list))
```

```python
Arr = [5,3,4,1,2,6]
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0,n-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr

print(bubble_sort(Arr))
```

```python
class Shape:
    def __init__(self,color,is_filled):
        self.color = color
        self.is_filled = is_filled
    def describe(self):
        print(f"{self.color},{self.is_filled}")
    def area():
        pass

class Reactangle(Shape):
    def __init__(self,color,is_filled,width,length):
        super().__init__(color,is_filled)
        self.width = width
        self.length = length
    def area():
        return f"Area of Reactangle :{self.length * self.width}"
class Circle(Shape):
    def __init__(self,color,is_filled,radius):
        super().__init__(color,is_filled,)
        self.radius = radius
    def area():
        return f"Area of Cirlce:{3.14* self.radius*self.radius}"

c = Circle('Green','True',5)
print(c.area())
        
```
